```python
# define a function to detect outliers
def outliers(df,ft):
    upperbond = 1
    ls = df.index[(df[ft] > upperbond)]
    
    return ls

# create a function to remove the outliers
def remove(df,ls):
    ls = sorted(set(ls))
    df = df.drop(ls,inplace=True)
    
    return df
```

```python
#create a list to store all the index of outliers
index_list = []
for feature in (['CASH_ADVANCE_FREQUENCY']):
    index_list.extend(outliers(dff, feature))
```

Principal Component Analysis (PCA) is a popular statistical method used for reducing the dimensionality of large datasets while retaining as much variance as possible. It is a technique used to transform a high-dimensional dataset into a lower-dimensional space while preserving the original variability of the dataset.

PCA works by identifying the directions, or principal components, in which the data varies the most. These principal components are linear combinations of the original variables, and each successive component accounts for as much of the remaining variance in the data as possible. The first principal component explains the maximum amount of variance in the data, the second principal component explains the maximum amount of remaining variance after accounting for the first component, and so on.

PCA is commonly used in data preprocessing and exploratory data analysis. It can help identify patterns in high-dimensional data that are difficult to discern visually, and it can be used to compress data into a lower-dimensional space, making it easier to visualize and work with. PCA is also used in machine learning and pattern recognition tasks as a tool for feature extraction and dimensionality reduction.

Normalization is important when working with CPA data

Reducing the impact of outliers: CPA data can sometimes contain outliers, which are data points that are significantly different from the other values in the data set. Outliers can have a significant impact on the analysis and can skew the results. Normalization can help reduce the impact of outliers by scaling the data to a common range.

Making data easier to interpret: Normalized data is easier to interpret because it is represented on a common scale. This makes it easier to compare different campaigns, advertisers, or time periods, and to identify trends and patterns in the data.

Facilitating machine learning and statistical analysis: Normalization is often a necessary preprocessing step when using machine learning algorithms or statistical analysis techniques. These techniques often require the data to be standardized and normalized to ensure that the results are accurate and meaningful.

Improving data visualization: Normalized data can be more easily visualized and plotted, which can help to identify patterns and trends in the data. Normalization can also help to reduce the scale of the data, making it easier to plot on a graph or chart.

in addition in this case since CPA maximizes variance, PCA is prone to select features with high variance 

```python
from sklearn.decomposition import PCA

# Create a PCA object with 3 components
pca = PCA(n_components=3)

# Fit the PCA model to the dataframe
pca.fit(normalized_df)

# Get the first three principal components
principal_components = pca.transform(normalized_df)[:, :3]

# Create a dataframe from the principal components
df_pca = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2', 'PC3'])

# Print the principal components dataframe
print(df_pca)
```

```python
variance_ratio = pca.explained_variance_ratio_
```

```python
variance_ratio
```

```python
suml=0
for i in range (len(variance_ratio)):
    suml= suml+variance_ratio[i]
    
    
print(suml)
    

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data_transformed = pca.transform(normalized_df)
```

```python
data_transformed
```

```python
inertia = []
for k in range(1, 10):
    # Initialize the KMeans model with random initialization
    kmeans = KMeans(n_clusters=k, init='random')
    
    # Fit the KMeans model to the transformed data
    kmeans.fit(data_transformed)
    
    # Add the inertia value to the list
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve of inertia
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
model_random = KMeans(n_clusters = 4, init = "random")
y_clusters_random = model_random.fit_predict(data_transformed)
```

```python
# countplot to check the number of clusters and number of customers in each cluster
sns.countplot(x = y_clusters_random)
```

Pros of performing PCA before clustering:

1.Dimensionality Reduction: PCA can help reduce the number of features or dimensions in the data while retaining the most important information. This can lead to improved clustering performance by reducing noise and focusing on the most relevant features.

2.Visualization: PCA can help visualize high-dimensional data by reducing it to two or three dimensions, which can be plotted on a graph. This can help in interpreting and understanding the data.

4.The training time of the algorithms reduces significantly with less number of features

5.Removes Correlated Features. From the density graph, we can see that the data before the PCA transformation still has some correlated features.

Disadvantages of PCA:

1.Information Loss: Pertain 77.87% of the original variance is good but it still means losing information.

2.Data standardization is must before PCA.

3.Difficulty in Interpreting Results: PCA can make it difficult to interpret the results of clustering, as the data has been transformed into a new space. This can make it harder to relate the clusters back to the original features or to understand the underlying patterns in the data. Independent variables become less interpretable, Principal Components are not as readable and interpretable as original features. We don't have the original features.
